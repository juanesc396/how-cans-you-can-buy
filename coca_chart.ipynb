```python
import matplotlib as mlp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tt = pd.read_csv("./saves/24-05-22.csv", encoding = 'cp1252')
tt.sort_values('cans', ascending = False, inplace = True)
tt['cans'] = round(tt['cans'])
tt.loc[15, 'country'] = 'Panamá'
tt.loc[15, 'city'] = 'Panamá'
```

```python
palette = ['#B1C0C4', '#8C0A07', '#AE2826', '#94090C', '#9C1714']

plt.figure(figsize = (18, 10))
bar = sns.barplot(x=tt['cans'], y=tt['country'], orient='h', palette = palette)
bar.set(ylabel=None)
bar.set_title('How many cans you can buy with 1000 ARS around the world?', fontsize=20, pad=15)
bar.bar_label(bar.containers[0], label_type='edge', padding = 6)

plt.savefig('./coke.png')
```
